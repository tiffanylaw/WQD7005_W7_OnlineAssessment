```python
#change date format
df['Date'] = pd.to_datetime(df.Date)
df
```

```python
#check change% is up/down
#then add value into new column indicator
df['indicator'] = '-'
```

```python
index = df.index
number_of_rows = len(index)
number_of_rows
```

```python
for n in range(number_of_rows):
    
    txt = df.loc[n]['Change %']
    x = re.findall("\-", txt)
    if(x):
        indicator = "down"
    else:
        indicator = "up"
        
    df['indicator'][n] = indicator
    
```
